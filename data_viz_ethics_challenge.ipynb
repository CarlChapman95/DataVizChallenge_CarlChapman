## Step 2: Review the BAD Charts

Below are two deliberately poor visualizations.  
Your task: Carefully review each and take notes on what you think is wrong, misleading, or just poorly designed.

```python
plt.figure(figsize=(8,5))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="yellow", linewidth=4, marker="o")
plt.xlabel("Year")
plt.ylabel("Total Sales")
#plt.ylim(subset["Sales"].min() * 0.98, subset["Sales"].min() * 1.02)  # extreme truncation
plt.title("Our Coffee Sales Over Time!", fontsize=16, fontweight="bold", color="darkblue")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()

#colors = ["red","orange","yellow","green","blue","purple"]

plt.bar(avg_sales["Region"], avg_sales["Sales"], 
        color='lightblue', edgecolor="black")#, linewidth=3, hatch="/", alpha=0.8)
plt.title("Coffee Sales by Region", fontsize=20, fontweight="bold", color="purple", backgroundcolor="yellow")
#plt.grid(True, which='both', color='black', linestyle='-', linewidth=1.5)
plt.xlabel("Regions", fontsize=16, fontweight="bold", color="black")
plt.ylabel("Sales", fontsize=16, fontweight="bold", color="black")
plt.xticks(rotation=45, fontsize=14, fontweight="bold", color="darkblue")
plt.yticks(fontsize=14, fontweight="bold", color="darkblue")
plt.show()
```

Now it’s your turn to fix things.

```python
# Your GOOD visualization(s) here
# Going to see how each region's sales over time compare to each other.


sales_over_time = coffee_df.groupby(['Year', 'Region'])['Sales'].sum().reset_index()

plt.figure(figsize=(12,6))
regions = sales_over_time['Region'].unique()
for region in regions:
    region_data = sales_over_time[sales_over_time['Region'] == region]
    plt.plot(region_data['Year'], region_data['Sales'])
    plt.text(region_data['Year'].values[0], region_data['Sales'].values[0], region)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Coffee Sales Over Time by Region')
plt.xticks(rotation=40)
plt.show()
```
